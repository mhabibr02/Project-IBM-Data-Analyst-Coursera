# Data Description

The dataset used for this visualization assignment contains *historical_automobile_sales* data representing automobile sales and related variables during recession and non-recession period.

<br>2. Recession: A binary variable indicating recession perion; 1 means it was recession, 0 means it was normal.

<br>5. Unemployment_Rate: The monthly unemployment rate.
<br>6. Consumer_Confidence: A synthetic index representing consumer confidence, which can impact consumer spending and automobile purchases.
<br>7. Seasonality_Weight: The weight representing the seasonality effect on automobile sales during the period.
<br>8. Price: The average vehicle price during the period.
<br>9. Advertising_Expenditure: The advertising expenditure of the company.
<br>10.Vehicle_Type: The type of vehicles sold; Supperminicar, Smallfamiliycar,                 Mediumfamilycar, Executivecar, Sports.
<br>11.Competition: The measure of competition in the market, such as the number of competitors or market share of major manufacturers.
<br>12.Month: Month of the observation extracted from Date..
<br>13.Year: Year of the observation extracted from Date.
<br>
By examining various factors mentioned above from the dataset, you aim to gain insights into how recessions impacted automobile sales for your company.

# Creating Visualizations for Data Analysis


### TASK 1.1
Develop a *Line chart* using the functionality of pandas to show how automobile sales fluctuate from year to year


````python
# Pastikan kolom yang digunakan sesuai dengan dataset
df_grouped = df.groupby("Year")["Automobile_Sales"].sum()

# Membuat grafik garis
plt.figure(figsize=(10, 5))
df_grouped.plot(kind="line", marker="o", linestyle="-", color="b")

# Menambahkan judul dan label
plt.title("Fluktuasi Penjualan Mobil per Tahun", fontsize=14)
plt.xlabel("Tahun", fontsize=12)
plt.ylabel("Total Penjualan Mobil", fontsize=12)
plt.grid(True)

# Menampilkan grafik
plt.show()
````

### TASK 1.2
Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?


````python
# Pastikan kolom yang digunakan sesuai dengan dataset
df_grouped = df.groupby(["Year", "Vehicle_Type"])["Automobile_Sales"].sum().unstack()

# Daftar tahun resesi berdasarkan data historis
recession_years = [2008, 2009]  # Bisa disesuaikan dengan data yang ada

# Membuat grafik
plt.figure(figsize=(12, 6))

# Plot garis untuk setiap jenis kendaraan
for column in df_grouped.columns:
    plt.plot(df_grouped.index, df_grouped[column], marker="o", linestyle="-", label=column)

# Menambahkan garis vertikal dan anotasi untuk tahun resesi
for year in recession_years:
    plt.axvline(x=year, color="r", linestyle="--", alpha=0.7)
    plt.text(year, df_grouped.max().max() * 0.9, "Recession", color="red", fontsize=12, ha="center")

# Menambahkan elemen grafis
plt.title("Sales Trends by Vehicle Type During Recession", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.legend(title="Vehicle Type")
plt.grid(True)

# Menampilkan semua tahun pada sumbu X
plt.xticks(ticks=df_grouped.index, labels=df_grouped.index, rotation=45)

# Menampilkan grafik
plt.show()
````

### TASK 1.3
The functionality of **Seaborn Library** to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.


````python
# Menentukan tahun resesi dan non-resesi
recession_years = [2008, 2009]
df["Recession"] = df["Year"].apply(lambda x: "Recession" if x in recession_years else "Non-Recession")

# Mengelompokkan data berdasarkan tahun, tipe kendaraan, dan kategori resesi
df_grouped = df.groupby(["Year", "Vehicle_Type", "Recession"])["Automobile_Sales"].sum().reset_index()

# Membuat plot menggunakan Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_grouped, x="Year", y="Automobile_Sales", hue="Vehicle_Type", style="Recession", markers=True, dashes=False)

# Menambahkan elemen visual
plt.title("Comparison of Vehicle Sales Trends: Recession vs. Non-Recession", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.legend(title="Vehicle Type")
plt.grid(True)

# Menampilkan grafik
plt.show()
````

### TASK 1.4
Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods?
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
# Daftar tahun resesi
recession_years = [2008, 2009]

# Pisahkan data berdasarkan periode
df_recession = df[df["Year"].isin(recession_years)]
df_non_recession = df[~df["Year"].isin(recession_years)]

# Membuat subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6), sharey=True)

# Plot untuk periode resesi
axes[0].plot(df_recession["Year"], df_recession["GDP"], marker="o", linestyle="-", color="red")
axes[0].set_title("GDP Trend During Recession", fontsize=14)
axes[0].set_xlabel("Year", fontsize=12)
axes[0].set_ylabel("GDP", fontsize=12)
axes[0].grid(True)

# Plot untuk periode non-resesi
axes[1].plot(df_non_recession["Year"], df_non_recession["GDP"], marker="o", linestyle="-", color="green")
axes[1].set_title("GDP Trend During Non-Recession", fontsize=14)
axes[1].set_xlabel("Year", fontsize=12)
axes[1].grid(True)

# Menampilkan grafik
plt.tight_layout()
plt.show()
````

### TASK 1.5
Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


````python
# Pastikan dataset memiliki kolom "Month"
df_grouped = df.groupby("Month")["Automobile_Sales"].sum().reset_index()

# Menyesuaikan ukuran gelembung berdasarkan penjualan
bubble_size = df_grouped["Automobile_Sales"] / df_grouped["Automobile_Sales"].max() * 100

# Membuat Bubble Plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_grouped, x="Month", y="Automobile_Sales", size="Automobile_Sales", sizes=(50, 500), alpha=0.6, color="blue", edgecolor="black")

# Menambahkan elemen visual
plt.title("Impact of Seasonality on Automobile Sales", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.xticks(ticks=range(1, 13), labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.grid(True)
plt.legend(title="Sales Volume", loc="upper right")

# Menampilkan grafik
plt.show()
````

 ### TASK 1.7
 Create a pie chart to display the portion of advertising expenditure of XYZAutomotives during recession and non-recession periods.
 <br>How did the advertising expenditure of XYZAutomotives change during recession and non-recession periods?


````python
# Menentukan tahun resesi
recession_years = [2008, 2009]

# Menghitung total pengeluaran iklan selama resesi dan non-resesi
ad_exp_recession = df[df["Year"].isin(recession_years)]["Advertising_Expenditure"].sum()
ad_exp_non_recession = df[~df["Year"].isin(recession_years)]["Advertising_Expenditure"].sum()

# Data untuk Pie Chart
labels = ["Recession", "Non-Recession"]
sizes = [ad_exp_recession, ad_exp_non_recession]
colors = ["red", "green"]

# Membuat Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=90, explode=(0.1, 0), shadow=True)

# Menambahkan judul
plt.title("Advertising Expenditure: Recession vs. Non-Recession", fontsize=14)

# Menampilkan grafik
plt.show()
````

### TASK 1.8
Develop a pie chart to display the total Advertisement expenditure for each vehicle type during recession period.<br>
Can we observe the share of each vehicle type in total expenditure during recessions?


````python
# Menentukan tahun resesi
recession_years = [2008, 2009]

# Filter data untuk periode resesi
df_recession = df[df["Year"].isin(recession_years)]

# Menghitung total pengeluaran iklan per jenis kendaraan
ad_exp_by_vehicle = df_recession.groupby("Vehicle_Type")["Advertising_Expenditure"].sum()

# Membuat Pie Chart
plt.figure(figsize=(8, 8))
# Remove the cmap argument
plt.pie(ad_exp_by_vehicle, labels=ad_exp_by_vehicle.index, autopct="%1.1f%%", startangle=90, shadow=True, colors=plt.cm.Set3.colors)

# Menambahkan judul
plt.title("Advertisement Expenditure by Vehicle Type (Recession)", fontsize=14)

# Menampilkan grafik
plt.show()
````

### TASK 1.10
Create a map on the hightest sales region/offices of the company during recession period


````python
# Menentukan tahun resesi
recession_years = [2008, 2009]

# Filter data untuk periode resesi
df_recession = df[df["Year"].isin(recession_years)]

# Menghitung total penjualan berdasarkan wilayah/kantor
sales_by_city = df_recession.groupby("City")[["Competition", "GDP", "Automobile_Sales"]].sum().reset_index()

# Menentukan lokasi pusat peta (misalnya, berdasarkan rata-rata koordinat)
map_center = [sales_by_city["Competition"].mean(), sales_by_city["GDP"].mean()]
sales_map = folium.Map(location=map_center, zoom_start=5)

# Menambahkan marker untuk setiap wilayah/kantor
for _, row in sales_by_city.iterrows():
    folium.CircleMarker(
        location=[row["Competition"], row["GDP"]],
        radius=row["Automobile_Sales"] / sales_by_city["Automobile_Sales"].max() * 15,  # Menyesuaikan ukuran lingkaran berdasarkan penjualan
        color="blue",
        fill=True,
        fill_color="blue",
        fill_opacity=0.6,
        popup=f"{row['City']}: {row['Automobile_Sales']} sales"
    ).add_to(sales_map)

# Menampilkan peta
sales_map
````
